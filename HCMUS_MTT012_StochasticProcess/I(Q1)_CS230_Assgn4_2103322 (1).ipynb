```python
'''
Name: Karan Sutariya
Roll No: 2103322
CS230 Assignment[4]: Part I - Q1 Simulating Bernoulli process and estimating the PMF of no. of tosses required till k-th HEAD comes up

'''


#Importing the following libraries to work with coin tossing and plotting the distribution


import numpy
from numpy.random import binomial
import matplotlib.pyplot as plot


k = 10

p = 0.5

PMF = {} #creating a dictionary to store the number of trials along with its count in 100000 iterations

for i in range(100000):
    
    #Initialization
    
    t = 0   # to keep a track of no. of trials
    no_heads = 0    # to keep a track of no. of Heads
    
    
    # tossing the coin tossing until we have obtained 'k' number of heads
    
    while ( no_heads != k ):
        
            outcome = binomial(1,p) #for developing coin tossing mechanism which gives 1( or H) with p probability and 0( or T) with (1-p) probability
            
            t = t + 1     #incrementing the value of t till it satisfies the condition
            
            if outcome != 0 : # iff the outcome is 1, the no. of heads are counted
                no_heads = no_heads + 1
    
      
    # Inserting the no. of trials and their respective counts in PMF
    if t in PMF:
        (PMF[t]) = (PMF[t]) + 1 #if already there then just add the counts
        
        
    else:
        PMF[t] =  1 
        
        
alpha = 100000 #the number of times these experiment is conducted for

#for plotting the probabilities, divide the values in dictionary by 100000
for k in PMF.keys():
    PMF[k] *= (1 / alpha)


#Plotting the distribution

plot.bar(PMF.keys(), PMF.values(), color = 'orange', ec='black')

plot.xlabel("Number of trials till 5th arrival/success  ------------->")

plot.ylabel("Probability  ------------->")

plot.title("EMPIRICAL PMF DISTRIBUTION")
```
